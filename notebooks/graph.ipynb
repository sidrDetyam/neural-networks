```python
import matplotlib.pyplot as plt
import numpy as np

min_dummy = 0
max_dummy = 10
attemps = 10
min_k = 3
max_k = 100
prefix = "results-kabby-lake/result"
#prefix = "results-nehalem/result"

plt.figure(figsize=(20,10))

results = []
for i in range(min_dummy, max_dummy+1):
    with open(f"{prefix}-{i}") as result:
        clocks = []
        local_clocks = []
        for j, line in enumerate(result):
            local_clocks.append(tuple(map(float, line.split(",")))[2])
            if (j+1) % attemps == 0:
                local_clocks.sort()
                clocks.append(min(local_clocks) / 100000)
                local_clocks.clear()

        results.append(clocks)

x = np.linspace(min_k, max_k, max_k-min_k+1)

for i in range(min_dummy, max_dummy+1):
    if(i % 1 == 0):
        plt.plot(x, results[i-min_dummy], label=f"dummy={i}")

plt.xticks(range(min_k, max_k, 3))
plt.xlabel('k')
plt.ylabel('clocks per cycle')

plt.legend()
plt.show()
```
